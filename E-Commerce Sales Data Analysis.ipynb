```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Transforming Data
df = pd.read_csv(r"D:\Project For Data Analyst\Data sets\OnlineRetail.csv\OnlineRetail.csv", encoding="cp1252")
print(df.head())
print(df.info())
print(df.describe())
df
```

```python
# Check Data Types
print(df.dtypes)
```

```python
# Now Change The Dtypes

df = df[df['InvoiceNo'].str.isnumeric()] #why am using this code because some of the 'invoiceno' having starts with alphabets thats the reason we are using this
df['InvoiceNo'] = df['InvoiceNo'].astype('int64')

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])


df['CustomerID'] = df['CustomerID'].astype('Int64')

df.dtypes
```

```python
# Checking If There Any Missing Values
print(df.isnull().any())

print(df.isnull().sum())

# Drop Duplicate Values
df = df.drop_duplicates()

# Identifying Missing Values Visually
df[df.isnull().any(axis=1)]


```

```python
# Check Missing Values Visually

import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.show()
```

```python
# Removing Null/Missing And values from the data set
df=df.dropna(subset=['CustomerID'])

# And Again Check The Missing Values
print(df.isnull().sum())
```

```python
# After Totally Cleaned Data Visually Checking If There Any Missing Values Or Duplicate Values

import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(12,5))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.show()

# Final Verification For Data Cleaned Or What
print(df.isna().sum())
print(df)
```

```python

# Remove canceled transactions
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

# Remove invalid quantities or prices

df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]

```

```python
# Create TotalSales 

df['TotalSales'] = df['Quantity'] * df['UnitPrice']
df
```

```python
# Extract date parts

df['Year'] = df['InvoiceDate'].dt.year

df['Month'] = df['InvoiceDate'].dt.month

df['MonthName'] = df['InvoiceDate'].dt.strftime('%b')

df
```

```python
# Optional – Create InvoiceMonth

df['InvoiceMonth'] = df['InvoiceDate'].dt.to_period('M')

# Exploratory Data Analysis (EDA)
# a) Basic summary
df.describe(include='all')

# b) Total revenue
total_sales = df['TotalSales'].sum()

# c) Top 10 products
top_products = (
    df.groupby('Description')['TotalSales']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

# d) Top 10 customers
top_customers = (
    df.groupby('CustomerID')['TotalSales']
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

# e) Sales by country
sales_country = (
    df.groupby('Country')['TotalSales']
      .sum()
      .sort_values(ascending=False)
)

# f) Monthly sales trend
monthly_sales = (
    df.groupby('InvoiceMonth')['TotalSales']
      .sum()
      .reset_index()
)
monthly_sales['InvoiceMonth'] = monthly_sales['InvoiceMonth'].dt.to_timestamp() #why iam using this na the monthly data type was in (period)datatype Seaborn cannot plot Period objects

# g) Average order value
order_value = df.groupby('InvoiceNo')['TotalSales'].sum().mean()

df.info()
df
```

```python
# Visualization

# Using Matplotlib / Seaborn:

import seaborn as sns
import matplotlib.pyplot as plt

# Monthly sales trend
plt.figure(figsize=(12,6))
sns.lineplot(x='InvoiceMonth', y='TotalSales', data=monthly_sales)
plt.title('Monthly Sales Trend')
plt.xticks(rotation=45)
plt.show()

# Top countries
plt.figure(figsize=(10,5))
top_countries = sales_country.head(10)
sns.barplot(x=top_countries.values, y=top_countries.index)
plt.title('Top 10 Countries by Revenue')
plt.show()
```
